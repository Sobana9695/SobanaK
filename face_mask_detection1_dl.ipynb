```python
data=[]  # feature
labels=[]  # label

for i in with_mask:
    img=cv2.imread(i)  # give array,append array to list
    img=cv2.resize(img,(250,250))
    data.append(img)
    labels.append(0)
    
for i in without_mask:
    img=cv2.imread(i)  
    img=cv2.resize(img,(250,250))
    data.append(img)
    labels.append(1)    
```
