```python
vid=cv2.VideoCapture(0)
count=0
while True:
    count=count+1
    su,img=vid.read()
    if su==True:
        cv2.imshow("video",img)
        images=cv2.imwrite(r"C:\Users\SSS\Desktop\opencv_task\1\{}.jpg".format(count),img)

    if cv2.waitKey(1)==ord("q"):
        break
cv2.destroyAllWindows()
```
